```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load every data sets into pandas dataframe
horse_df = pd.read_csv('datasets/part2/horseasses-population-in-nepal-by-district.csv')
milk_df = pd.read_csv('datasets/part2/milk-animals-and-milk-production-in-nepal-by-district.csv')
meat_df = pd.read_csv('datasets/part2/net-meat-production-in-nepal-by-district.csv')
cotton_df = pd.read_csv('datasets/part2/production-of-cotton-in-nepal-by-district.csv')
egg_df = pd.read_csv('datasets/part2/production-of-egg-in-nepal-by-district.csv')
rabbit_df = pd.read_csv('datasets/part2/rabbit-population-in-nepal-by-district.csv')
wool_df = pd.read_csv('datasets/part2/wool-production-in-nepal-by-district.csv')
yak_df = pd.read_csv('datasets/part2/yak-nak-chauri-population-in-nepal-by-district.csv')
```

```python
# change meat dataframe's "SANKHUWASHABHA" to "SANKHUWASHAVA" for matching purpose
meat_df.iloc[1,0] = 'SANKHUWASHAVA'
```

```python
# removing space in region -> E. Region to E.Region
milk_df.iloc[19,0] = 'E.REGION'
milk_df.iloc[42,0] = 'C.REGION'
milk_df.iloc[62,0] = 'W.REGION'
milk_df.iloc[81,0] = 'MW.REGION'
milk_df.iloc[94,0] = 'FW.REGION'
```

```python
# start data merging
merged_df = egg_df.merge(milk_df, on='DISTRICT')
merged_df = merged_df.merge(meat_df, on='DISTRICT')
merged_df = merged_df.merge(wool_df, on='DISTRICT')
```

```python
pd.set_option('display.max_rows', None)
merged_df = merged_df.merge(yak_df, on='DISTRICT', how='left')
merged_df = merged_df.merge(horse_df, on='DISTRICT', how='left')
merged_df = merged_df.merge(rabbit_df, on='DISTRICT', how='left')
```

```python
# update district column name to uppercase
cotton_df['DISTRICT'] = cotton_df['DISTRICT'].str.upper()
```

```python
merged_df = merged_df.merge(cotton_df, on='DISTRICT', how='left')
```

```python
# changing NAN values to zero
final_df = merged_df.replace(np.nan,0)
```

## EDA Start

```python
regions = ['E.REGION','C.REGION', 'W.REGION', 'MW.REGION','FW.REGION']
"""
E.REGION -> Eastern region
C.REGION -> Central region
W.REGION -> Western region
MW.REGION -> Mid-Western region
FW.REGION -> Far-Western region
"""
        
horse_by_region = horse_df.loc[horse_df['DISTRICT'].isin(regions)] 
milk_by_region = milk_df.loc[milk_df['DISTRICT'].isin(regions)] 
meat_by_region = meat_df.loc[meat_df['DISTRICT'].isin(regions)] 
egg_by_region = egg_df.loc[egg_df['DISTRICT'].isin(regions)] 
rabbit_by_region = rabbit_df.loc[rabbit_df['DISTRICT'].isin(regions)] 
wool_by_region = wool_df.loc[wool_df['DISTRICT'].isin(regions)] 
yak_by_region = yak_df.loc[yak_df['DISTRICT'].isin(regions)] 
```

## Milk production data viz

```python
# group bar graph for milk production by regions
X_axis = np.arange(len(milk_by_region['DISTRICT']))
X = milk_by_region['DISTRICT'].values
buffalo_milk = milk_by_region['BUFF MILK'].values
cow_milk = milk_by_region['COW MILK'].values

plt.bar(X_axis-0.2, buffalo_milk, 0.4, label="Buffalo milk")
plt.bar(X_axis+0.2, cow_milk, 0.4, label="Cow milk")
plt.xlabel('Milk production by region')
plt.ylabel("Milk production in ltr")
plt.xticks(X_axis, X)
plt.xticks(X_axis,milk_by_region['DISTRICT'] )
plt.yticks([50000, 100000, 150000,200000, 250000, 300000,350000], ['50k','100k','150k','200k','250k','300k','350k'])
plt.legend()
plt.show()
```

## Horses population pie chart

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(horse_by_region['Horses/Asses'].values,labels=horse_by_region['DISTRICT'].values)
plt.legend()
plt.plot()
```
